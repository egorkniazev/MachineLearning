```python
import matplotlib.pyplot as plt
# from kneed import KneeLocator
from sklearn.datasets import make_blobs, make_moons
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
```

```python
# features, true_labels = make_blobs(
#     n_samples=200,
#     centers=3,
#     cluster_std=2.75,
#     random_state=42
# )

features, true_labels = make_moons(n_samples=1500, 
                                   shuffle=True,
                                   noise=0.01)
```

```python
features[:5]
```

```python
true_labels[:5]
```

```python
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

```python
scaled_features[:5]
```

```python
kmeans = KMeans(
    init="random",
    n_clusters=2,
    n_init=10,
    max_iter=300,
    random_state=42
)
```

```python
kmeans.fit(scaled_features)
```

```python
# The lowest SSE value
kmeans.inertia_
```

```python
# Final locations of the centroid
kmeans.cluster_centers_
```

```python
# The number of iterations required to converge
kmeans.n_iter_
```

```python
kmeans.labels_[:5]
```

```python
plt.scatter(features[:, 0], features[:, 1], c=kmeans.predict(features))
```

```python
features, true_labels = make_blobs(
    n_samples=2000,
    centers=4,
    cluster_std=2.75,
    random_state=42
)
```

```python
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

```python
kmeans = KMeans(
    init="random",
    n_clusters=4,
    n_init=10,
    max_iter=300,
    random_state=42
)
```

```python
kmeans.fit(scaled_features)
```

```python
plt.scatter(features[:, 0], features[:, 1], c=kmeans.predict(features))
```
